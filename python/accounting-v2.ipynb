```python
import requests, json
import csv
import io,os
from datetime import datetime
import time


#intervals must be 365 days or less
#intervals =[["2022-01-01","2022-12-31"],["2023-01-01","2023-12-31"],["2024-01-01","2024-12-31"],["2025-01-01","2025-12-31"]]
intervals = [["2025-01-01","2025-12-31"]]


#all the keys
keys=[
"5GL46po3iBatb3uyWabqy7tUeKK6Wu8HFmu4rj1z2Pmu7VbV" 
]



api_key= "<your API KEY>"

headers = {
    "accept": "application/json",
    "Authorization": api_key
}
```

```python
#loop through the keys
for key in keys:

    #loop through the intervals
    for interval in intervals:
 
        #tao is default staking
        #find subnets that had staking events
        start_time = interval[0]
        end_time = interval[1]
        year = start_time[:4]
        url = f"https://api.taostats.io/api/accounting/tax_token/v1?date_start={start_time}&date_end={end_time}&coldkey={key}"

        response = requests.get(url, headers=headers)
        resJson = json.loads(response.text)
        #all the tokens in the interval
        tokens = resJson['data']
        #loop through all the tokens held for a key in an interval
        for token in tokens:
            url = f"https://api.taostats.io/api/accounting/tax_csv/v1?token={token}&date_start={start_time}&date_end={end_time}&coldkey={key}"
            response = requests.get(url, headers=headers)
            
            #resJson = json.loads(response.text)
            acountingcsv = response.text
            name = f"{key}_{year}_{token}_accounting"
            print(name)
            with open(f'{name}.csv', 'w', newline='') as f:
                writer = csv.writer(f)
                reader = csv.reader(io.StringIO(acountingcsv), delimiter=',', quotechar='"')
                for row in reader:
                    writer.writerow(row)
               
                
    
```
