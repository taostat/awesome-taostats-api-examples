```python
from tabulate import tabulate

def emission_calculator(blocks, tao_in, alpha_in, emission, price):
    
    price = min(emission, price)
    
    initial_price = price
    print("initial price", price)
    alpha_to_root = 0.41
    
    tao_in_block = min(emission, price)
    alpha_in_block = min(1, tao_in/price)
    tao_sold = 0
    price_delta = price/initial_price*100
    counter = 1
    
    
    data = []
    data.append([counter,tao_in, tao_in_block, alpha_in, alpha_in_block, tao_sold, price, price_delta])
    
    while counter <= num_of_blocks:
        
        #emissions
        tao_in_block = min(emission, price)
        alpha_in_block = min(1, tao_in/price)
        alpha_out_block = 2- alpha_in_block
        
        #sell 100% of vali emission to root
        vali_alpha = alpha_out_block*.41
    
        
    
        k = alpha_in * tao_in
    
        #decrease tao_in by the amount of tao sold
    
        tao_sold = tao_in - k/(alpha_to_root + alpha_in)
    
        tao_in-= tao_sold
    
        #increase alpha in
        alpha_in += vali_alpha
        alpha_in += alpha_in_block
    
        #increase emission on tao_in
        tao_in += tao_in_block
        
        price = tao_in/alpha_in
        price_delta = price/initial_price*100
        counter +=1
        data.append([counter, tao_in, tao_in_block, alpha_in, alpha_in_block, tao_sold, price, price_delta])
    
    
    print(f"price after {counter} blocks: {price}")
    print(f"{round(price/initial_price*100,2)}% of initial price")
    print(tabulate(data, headers=["Block","Tao In", "Tao In Block", "Alpha In", "Alpha In Block", "Tao Sold", "Price", "Price %"], tablefmt="grid"))

    
```

```python
num_of_blocks = 7200

tao_in = 1
alpha_in =64
price = tao_in/alpha_in
emission = tao_in/alpha_in

emission_calculator(num_of_blocks, tao_in, alpha_in, emission, price)
```
